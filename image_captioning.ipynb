```python
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
mylist = []
```

```python
def create_ds(starts: int = 0):
    # x_9gag_scrapper\Image
    img_path = "./x_9gag_scrapper/Image"
    
    images = sorted(os.listdir(img_path), key=lambda x: int(x.split('.')[0]))
    
    for image in images:
        # print(f"{img_path}/{image}")
        base64_image = encode_image(f"{img_path}/{image}")
        caption = client.chat.completions.create(
        model="llama-3.2-90b-vision-preview",
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": "Describe this meme shortly in 1-2 sentences and how it represents Islam. Give the answer in plain text and not markdown. No paragraphs"
                    },
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": f"data:image/jpeg;base64,{base64_image}",
                        }
                    },
                ]
            }
        ],
            temperature=0,
            max_tokens=512,
        )
        content = caption.choices[0].message.content
        mylist.append(content)
        print(content)
```
